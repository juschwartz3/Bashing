## Integrate Pie Charts onto Map

### Subtask:
Create a single Altair chart that generates pie charts at the geographical coordinates of Ivory Coast and Colombia, representing the allele frequencies of the two samples, and then layer this chart onto the base world map.


**Reasoning**:
To create the pie charts for each location and layer them onto the world map, I need to define an Altair chart that transforms the allele frequency data into angular segments suitable for pie charts. This involves calculating start and end angles for each allele type within each sample, and then encoding these values to create the arc marks, positioned by latitude and longitude.



```python
import altair as alt
import math

# 1. Define a new Altair chart named pie_charts_layer using the pie_chart_locations_data DataFrame.
pie_charts_layer = alt.Chart(pie_chart_locations_data).transform_window(
    # 2. Add a transform_window to this chart to calculate startAngle and endAngle for the arc segments.
    theta="sum(Allele Frequency)",
    stack=True, # This helps to correctly stack the angles for the pie chart
    groupby=['Sample'],
    sort=[alt.EncodingSortField('Allele Type', op='min', order='descending')], # Sort to ensure consistent stacking order
    frame=[None, None]
).transform_calculate(
    # Calculate startAngle and endAngle based on the stacked theta
    startAngle="(datum.theta - datum['Allele Frequency']) * 2 * PI",
    endAngle="datum.theta * 2 * PI"
).mark_arc(
    # 3. Use mark_arc() to draw the pie slices.
    # 7. Set the outerRadius of the arcs and apply a stroke with a strokeWidth.
    outerRadius=15,
    stroke='black',
    strokeWidth=0.5
).encode(
    # 4. Encode the longitude and latitude from pie_chart_locations_data to position each pie chart on the map.
    longitude='Longitude:Q',
    latitude='Latitude:Q',
    # 5. Encode theta using startAngle and endAngle
    theta=alt.Theta(field="Allele Frequency", type="quantitative", stack=True), # Use AF for theta to represent proportion
    # 6. Encode color by 'Allele Type' to differentiate between resistance and reference alleles.
    color=alt.Color('Allele Type:N', title='Allele Type'),
    # 9. Add tooltips for 'Sample', 'Allele Type', and 'Allele Frequency' (formatted as percentage).
    tooltip=['Sample:N', 'Allele Type:N', alt.Tooltip('Allele Frequency:Q', format='.1%', title='Allele Frequency')]
).project(
    # 8. Apply the same project method (type='mercator') to this pie_charts_layer as was used for the base_map.
    type='mercator'
)

# 10. Finally, combine the base_map and the pie_charts_layer using alt.layer() and display the resulting layered chart.
final_map_chart = alt.layer(base_map, pie_charts_layer).properties(
    title='Allele Frequencies at AA Position 108 by Location'
)

final_map_chart
```

```python
import altair as alt
import math

# 1. Define a new Altair chart named pie_charts_layer using the pie_chart_locations_data DataFrame.
pie_charts_layer = alt.Chart(pie_chart_locations_data).transform_window(
    # This transform_window calculates the cumulative sum of 'Allele Frequency' within each sample.
    frame=[None, 0], # For cumulative sum
    groupby=['Sample'],
    sort=[alt.EncodingSortField('Allele Type', order='descending')], # Sort to ensure consistent pie slices
    cumulative_af="sum(Allele Frequency)" # Define a new field for the cumulative allele frequency
).transform_calculate(
    # Calculate startAngle and endAngle based on the cumulative_af
    endAngle=f"datum.cumulative_af * 2 * {math.pi}",
    startAngle=f"(datum.cumulative_af - datum['Allele Frequency']) * 2 * {math.pi}"
).mark_arc(
    # Use mark_arc() to draw the pie slices.
    # Set the outerRadius of the arcs and apply a stroke with a strokeWidth.
    outerRadius=15,
    stroke='black',
    strokeWidth=0.5
).encode(
    # Encode the longitude and latitude from pie_chart_locations_data to position each pie chart on the map.
    longitude='Longitude:Q',
    latitude='Latitude:Q',
    # Encode theta using startAngle and endAngle
    theta=alt.Theta(field="endAngle", type="quantitative", stack="zero"), # As per instruction example
    theta2=alt.Theta(field="startAngle"),
    # Encode color by 'Allele Type' to differentiate between resistance and reference alleles.
    color=alt.Color('Allele Type:N', title='Allele Type'),
    # Add tooltips for 'Sample', 'Allele Type', and 'Allele Frequency' (formatted as percentage).
    tooltip=['Sample:N', 'Allele Type:N', alt.Tooltip('Allele Frequency:Q', format='.1%', title='Allele Frequency')]
).project(
    # Apply the same project method (type='mercator') to this pie_charts_layer as was used for the base_map.
    type='mercator'
)

# Filter selected_sample_locations for Ivory Coast to get its coordinates
ivory_coast_coords = selected_sample_locations[selected_sample_locations['Location'] == 'Ivory Coast'].iloc[0]
colombia_coords = selected_sample_locations[selected_sample_locations['Location'] == 'Colombia'].iloc[0]

# Create a text layer for Ivory Coast annotation
ivory_coast_text = alt.Chart(pd.DataFrame({
    'Location': ['Ivory Coast'],
    'Latitude': [ivory_coast_coords['Latitude'] + 5],  # Adjusted latitude offset further
    'Longitude': [ivory_coast_coords['Longitude']],
    'Text': ['Ivory Coast']
})).mark_text(align='center', baseline='bottom', dy=-10).encode(
    longitude='Longitude:Q',
    latitude='Latitude:Q',
    text='Text:N',
    color=alt.value('black')
).project(
    type='mercator'
)

# Create a text layer for Colombia annotation
colombia_text = alt.Chart(pd.DataFrame({
    'Location': ['Colombia'],
    'Latitude': [colombia_coords['Latitude'] + 5], # Adjusted latitude offset further
    'Longitude': [colombia_coords['Longitude']],
    'Text': ['Colombia']
})).mark_text(align='center', baseline='bottom', dy=-10).encode(
    longitude='Longitude:Q',
    latitude='Latitude:Q',
    text='Text:N',
    color=alt.value('black')
).project(
    type='mercator'
)


# Finally, combine the base_map, the pie_charts_layer and text annotations using alt.layer() and display the resulting layered chart.
final_map_chart = alt.layer(base_map, pie_charts_layer, ivory_coast_text, colombia_text).properties(
    title='Allele Frequencies at AA Position 108 by Location',
    width=800,  # Increased width
    height=500  # Increased height
)

final_map_chart
```
